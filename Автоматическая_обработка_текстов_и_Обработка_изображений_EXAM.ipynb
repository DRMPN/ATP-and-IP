# **NLP**

```python
import numpy as np
def PPMI(N, c1, c2, c_both):
  return max([np.log2(N*c_both/c1/c2) ,0])
```

```python
import pandas as pd
data = pd.read_csv("/content/text.csv", header = None)
data.head()
```

```python
Length = len(data[0])
Length
```

```python
pair1_1 = "people"
pair1_2 = "closer"

pair2_1 = "first"
pair2_2 = "look"

pair3_1 = "article"
pair3_2 = "first"

c1_both, c11, c12, c2_both, c21, c22, c3_both, c31, c32 = 0, 0, 0, 0, 0, 0, 0, 0, 0
```

```python
for i in range(Length):
  if pair1_1 in data[0][i]:
    c11+=1
  if pair1_2 in data[0][i]:
    c12+=1
  if pair1_1 in data[0][i] and pair1_2 in data[0][i]:
    c1_both+=1

  if pair2_1 in data[0][i]:
    c21+=1
  if pair2_2 in data[0][i]:
    c22+=1
  if pair2_1 in data[0][i] and pair2_2 in data[0][i]:
    c2_both+=1
    
  if pair3_1 in data[0][i]:
    c31+=1
  if pair3_2 in data[0][i]:
    c32+=1
  if pair3_1 in data[0][i] and pair3_2 in data[0][i]:
    c3_both+=1
```

```python
print(f'Answer for pair: {pair1_1} and {pair1_2} is', PPMI(Length, c11, c12, c1_both))
print(f'Answer for pair: {pair2_1} and {pair2_2} is', PPMI(Length, c21, c22, c2_both))
print(f'Answer for pair: {pair3_1} and {pair3_2} is', PPMI(Length, c31, c32, c3_both))
```
